对10个数进行排序。

```python
lst = list(range(10, 0, -1))
```

```python
for j in range(len(lst),0,-1):
    for i in range(j-1):
        if lst[i] > lst[i+1]:
            lst[i], lst[i+1] = lst[i+1], lst[i]
```

```python
lst_endpos = len(lst) - 1
for i in range(lst_endpos):
    for j in range(0, lst_endpos - i):
        if lst[j] > lst[j + 1]:
            lst[j], lst[j+1] = lst[j+1], lst[j]
```
